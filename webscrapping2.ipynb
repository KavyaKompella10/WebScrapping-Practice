```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://editorial.rottentomatoes.com/guide/best-netflix-movies-to-watch-right-now/'
response = requests.get(url)
print(response)
```

```python
soup = BeautifulSoup(response.text, 'html')
# print(soup)
```

```python
movie_containers = soup.find_all('div', class_='row countdown-item')


for container in movie_containers:
    title_tag = container.find('div', class_='article_movie_title')
    title = title_tag.find('a').get_text() if title_tag else 'N/A'
    year_tag = title_tag.find('span', class_='subtle start-year') if title_tag else None
    year = year_tag.get_text().strip('()') if year_tag else 'N/A'
    score_tag = container.find('span', class_='tMeterScore')
    score = score_tag.get_text() if score_tag else 'N/A'
    consensus_tag = container.find('div', class_='info critics-consensus')
    consensus = consensus_tag.get_text(strip=True).replace('Critics Consensus:', '') if consensus_tag else 'N/A'
    synopsis_tag = container.find('div', class_='info synopsis')
    synopsis = synopsis_tag.get_text(strip=True).replace('Synopsis:', '') if synopsis_tag else 'N/A'
    print(f'Title: {title}')
    print(f'Year: {year}')
    print(f'Score: {score}')
    print(f'Critics Consensus: {consensus}')
    print(f'Synopsis: {synopsis}')
    print("\n")
```
